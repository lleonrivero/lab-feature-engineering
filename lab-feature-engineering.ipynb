## Use appropriate methods to clean the columns GEOCODE2, WEALTH1, ADI, DMA,and MSA.

#### GEOCODE2

```python
print(categorical['GEOCODE2'].value_counts())
print(categorical['GEOCODE2'].isna().sum())
```

```python
# Cleaning GEOCODE2. I will use the mode.
# It contains both NaNs and cells with spaces. Two different steps are required. 

# Function to replace cells with spaces with the mode
def clean_geocode(x):
    
    if x == ' ':
        return 'A'
    else:
        return x

categorical['GEOCODE2'] = list(map(clean_geocode, categorical['GEOCODE2']))
```

```python
# Replacing NaNs with the mode
categorical['GEOCODE2'] = categorical['GEOCODE2'].fillna('A')
```

```python
print(categorical['GEOCODE2'].value_counts())
print(categorical['GEOCODE2'].isna().sum())
```

#### WEALTH1

```python
print(numerical['WEALTH1'].value_counts())
print(numerical['WEALTH1'].isna().sum()/len(numerical['WEALTH1'])*100)
```

```python
# Cleaning WEALTH1.
# The column contains roughly 50% missing values.
# All the other values are reasonably well balanced.
# I will drop the column

numerical = numerical.drop(['WEALTH1'], axis = 1)
```

#### ADI

```python
print(numerical['ADI'].isna().sum())
print(numerical['DMA'].isna().sum())
print(numerical['MSA'].isna().sum())
```

```python
# Interestingly, these three columns contain the exact same amount of NaNs. 
# Checking whether the rows are the same. If so, I will be dropping them.
```

```python
adi_dma = numerical[(numerical['ADI'].isna())] == numerical[(numerical['DMA'].isna())]
```

```python
adi_dma_msa = adi_dma == numerical[(numerical['MSA'].isna())]
```

```python
adi_dma_msa['ADI'].shape
```

```python
# The rows with missing values seem to be the same in ADI, DMA and MSA. I will then be dropping them.
numerical = numerical[numerical['ADI'].isna()==False]
```

```python
# Now these three columns have 0 NaNs.
print(numerical['ADI'].isna().sum())
print(numerical['DMA'].isna().sum())
print(numerical['MSA'].isna().sum())
```
